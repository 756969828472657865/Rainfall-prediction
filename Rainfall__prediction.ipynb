```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Load the dataset
file_path = '/content/rainfall_prediction_dataset.csv'
df = pd.read_csv(file_path)

# Basic Checks
print("\nDataset Info:\n")
print(df.info())
print("\nMissing Values:\n")
print(df.isnull().sum())
print("\nStatistical Summary:\n")
print(df.describe())

# Exploratory Data Analysis (EDA)
sns.histplot(df['PrecipitationSumInches'], bins=30, kde=True)
plt.title('Precipitation Distribution')
plt.show()

# Boxplot for Outlier Detection
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['TempHighF', 'TempLowF', 'HumidityHighPercent', 'HumidityLowPercent']])
plt.title('Boxplot for Temperature and Humidity Features')
plt.show()

# Correlation Heatmap
# Select only numerical features for correlation calculation
numerical_features = df.select_dtypes(include=np.number)

sns.heatmap(numerical_features.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Feature Correlation Heatmap')
plt.show()


# Time Series Visualization for Rainfall Trends
df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(14, 6))
plt.plot(df['Date'], df['PrecipitationSumInches'])
plt.title('Rainfall Over Time')
plt.xlabel('Date')
plt.ylabel('Precipitation (Inches)')
plt.show()

# Feature Engineering
# Delete irrelevant or low-variance features if any (e.g., columns with same value)
df = df.drop(columns=[col for col in df.columns if df[col].nunique() == 1])

# Create new features (Example: Temperature Range)
df['TempRangeF'] = df['TempHighF'] - df['TempLowF']

# Convert 'Events' column to numeric using Label Encoding
label_encoder = LabelEncoder()
df['Events'] = label_encoder.fit_transform(df['Events'])

# Define target variable: Predict if precipitation occurs (1 if > 0, else 0)
df['PrecipitationBinary'] = df['PrecipitationSumInches'].apply(lambda x: 1 if x > 0 else 0)

# Prepare features and target without balancing
features = df.drop(['Date', 'PrecipitationSumInches', 'PrecipitationBinary'], axis=1)
target = df['PrecipitationBinary']

# Normalize the feature data
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=0.2, random_state=42)

# Model Implementation: K-Nearest Neighbors (KNN)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# Model Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"Initial KNN Model Accuracy: {accuracy:.2f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.show()

# Neural Network Implementation (MLP)
nn_model = Sequential()
nn_model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
nn_model.add(Dense(32, activation='relu'))
nn_model.add(Dense(1, activation='sigmoid'))

# Compile the model
nn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
nn_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=1)

# Evaluate Neural Network
loss, nn_accuracy = nn_model.evaluate(X_test, y_test)
print(f"Neural Network Model Accuracy: {nn_accuracy:.2f}")

# Linear Regression Implementation
reg_model = LinearRegression()
reg_model.fit(X_train, y_train)
reg_predictions = reg_model.predict(X_test)

# Regression Model Evaluation
mse = mean_squared_error(y_test, reg_predictions)
r2 = r2_score(y_test, reg_predictions)
print(f"\nLinear Regression MSE: {mse:.2f}")
print(f"Linear Regression R^2 Score: {r2:.2f}")

# Single Layer Perceptron
slp_model = Sequential()
slp_model.add(Dense(1, input_dim=X_train.shape[1], activation='sigmoid'))

# Compile the model
slp_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
slp_model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=1)

# Evaluate Single Layer Perceptron
slp_loss, slp_accuracy = slp_model.evaluate(X_test, y_test)
print(f"Single Layer Perceptron Accuracy: {slp_accuracy:.2f}")

# Hyperparameter Tuning
param_grid = {
    'n_neighbors': list(range(1, 21)),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

print(f"\nBest Parameters: {grid_search.best_params_}")

# Retrain and Re-evaluate
best_knn = grid_search.best_estimator_
y_pred_best = best_knn.predict(X_test)
accuracy_best = accuracy_score(y_test, y_pred_best)
print(f"\nTuned KNN Model Accuracy: {accuracy_best:.2f}")
print("\nClassification Report:\n", classification_report(y_test, y_pred_best))

# Model Comparison
print(f"Initial KNN Model Accuracy: {accuracy:.2f}")
print(f"Tuned KNN Model Accuracy: {accuracy_best:.2f}")
print(f"Neural Network Model Accuracy: {nn_accuracy:.2f}")
print(f"Single Layer Perceptron Accuracy: {slp_accuracy:.2f}")
print(f"Linear Regression R^2 Score: {r2:.2f}")
```

```python
# Model Comparison with Best Model Selection
model_accuracies = {
    'Initial KNN': accuracy,
    'Tuned KNN': accuracy_best,
    'Neural Network': nn_accuracy,
    'Single Layer Perceptron': slp_accuracy,
    'Linear Regression R^2': r2
}

# Plotting Model Accuracies
plt.figure(figsize=(10, 6))
model_names = list(model_accuracies.keys())
accuracy_scores = list(model_accuracies.values())
sns.barplot(x=accuracy_scores, y=model_names, palette='viridis')
plt.title('Model Performance Comparison')
plt.xlabel('Accuracy / R^2 Score')
plt.ylabel('Models')
plt.show()

# Best Model Selection
best_model_name = max(model_accuracies, key=model_accuracies.get)
print(f"\nThe best model based on evaluation is: {best_model_name} with a score of {model_accuracies[best_model_name]:.2f}")
```
